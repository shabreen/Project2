```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
sap500_df = pd.read_csv("datasets/SAP500.csv")
```

```python
sap500_df.head()
```

```python
#create connection to sqlite
engine = create_engine('sqlite:///db/BullsBearsDB.sqlite', echo=False)
```

```python
#export scraped data to sqlite
sap500_df.to_sql('SAP500', con=engine, if_exists='replace', index=False)
```

```python
#check if data was read into sqlite
engine.execute("SELECT * FROM SAP500").fetchall()
```
